```python
data = np.array([25, 37, 24, 28, 28, 35, 22, 31, 53, 41, 64, 29, 120, 72])
```

## Using Statistical Library

```python
# Mean
data.mean()
```

```python
? np.percentile
```

```python
#Median
np.percentile(data, 50.0)
```

```python
? stats.mode
```

```python
#Mode
stats.mode(data)[0]
```

```python
# Generate Plot to Visualize

import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(15,8)) #template size

ax=sns.distplot(data, kde=True)
```

```python
? sns.distplot
```

```python
plt.figure(figsize=(15,8)) #template size

ax=sns.distplot(data, kde=True)
plt.axvline(x=data.mean(), linewidth=3, color='g', label="mean", alpha=0.5)
plt.axvline(x=np.percentile(data, 50.0), linewidth=3, color='y', label="median", alpha=0.5)
plt.axvline(x=stats.mode(data)[0], linewidth=3, color='b', label="mode", alpha=0.5)
plt.legend(["Distribution","Mean", "Median", "Mode"])
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import math
```

```python
raw_data = [20,30,40,30]
friends_salary = np.array(raw_data)

print(np.ptp(friends_salary)) #Range of values in Numpy
print(np.var(friends_salary)) # Variance
print(np.std(friends_salary)) # Standard Deviation
print(stats.iqr(friends_salary)) # IQR p75 - p25
```

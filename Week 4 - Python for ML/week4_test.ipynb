```python
# Plot a categorical variable

names = ['group a', 'group b', 'group c']
values = [1,10,100]

#1
plt.bar(names, values)
plt.show()

#2
plt.figure(figsize=(9,3))
plt.bar(names, values)
plt.show()

#3.1
plt.subplot(131)
plt.bar(names, values)

#3.2
plt.subplot(132)
plt.scatter(names, values)

#3.3
plt.subplot(133)
plt.plot(names, values) # default is line plot

plt.suptitle("Categorical plots")

plt.show()
```

**SEABORN**

```python
import seaborn as sns
import pandas as pd

iris = sns.load_dataset("iris")
print(iris.head())

type(iris)
```

```python
plt.scatter(iris["sepal_length"], iris["sepal_width"])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Plot")
plt.show()
```

```python
# Box Plot

plt.boxplot(x = 'petal_length', data=iris) # 25 and 75th quartile IQR and IDR similiar
plt.show()
```

```python
# Histogram

plt.hist(iris['petal_width'], bins=100)
plt.show()
```

```python
# Pie chart

species_count = iris["species"].value_counts()
plt.pie(species_count, labels=species_count.index)
plt.show()
```

```python
sns.set_theme(style='ticks')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
sns.set_palette(sns.color_palette("pastel"))

fig,ax = plt.subplots(nrows=2, ncols=2, figsize = (15,7))

sns.boxplot(x = 'sepal_length', data=iris, hue='species', ax=ax[0][0])
sns.violinplot(x='species', y='petal_length', data=iris, ax=ax[0][1])
sns.histplot(y='petal_length',bins=10, data=iris, hue='species',ax=ax[1][0])
sns.kdeplot(x='petal_width', data=iris, hue='species',ax=ax[1][1])

fig.suptitle("4 plots at once")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lmplot(x='petal_length', y='sepal_length', data=iris, hue='species')
plt.show()
```

```python
sns.heatmap(iris.corr(), annot=True )
plt.show()
```

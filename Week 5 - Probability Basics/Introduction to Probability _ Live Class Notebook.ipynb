# Probability Introduction

When Dr. Strange said we have 1 in 14 million chances of winning the war, he practically saw those 14 million futures, this is called frequentist probability, which defines an event's probability as the limit of its relative frequency in a large number of trials. But not always do we have Dr. Strange's time stone to see all the possible futures or events that are repeatable, in this case we turn to Bayesian probability, which uses probability to represent a degree of belief for certain events, with 1 indicating absolute certainty and 0 indicating absolute uncertainty.

Even though the frequentist probability is related to rates at which events occur and Bayesian probability is related to qualitative levels of certainty, we treat both of them as behaving the same and we use the exact same formulas to compute the probability of events.

Probability theory is a mathematical framework for representing uncertain statements. But probability is not just an abstract concept in the math world, probability is all around us, and it can be fun to calculate the probability of events in our life.

For example, the next time you get late in traffic and your boss calls you, in addition to the time you will reach, you can also say the probability of you reaching in that time, of course it would take you a little bit of math but if you did miss a meeting, you can tell your boss you did tell him there was a chance you might not make it right 😉.

In artificial intelligence applications, we use probability theory in two major ways.

First, the laws of probability tell us how AI systems should reason, so we design our algorithms to compute or approximate various expressions derived using probability theory.

Second, we can use probability and statistics to theoretically analyze the behavior of proposed AI systems.

While probability theory allows us to make uncertain statements and to reason in the presence of uncertainty, information theory enables us to quantify the amount of uncertainty in a probability distribution.

Here, we are conduction the experiment with 'n' number of trial by defining a probability function.

```python
def perform_experiment(experiment, n):
    """Performs an exeriment n times and returns a list of results"""
    results = []
    for _ in range(n):
        result = experiment()
        results.append(result)
    return results
```

Counts the number of results matching a given condition

```python
def count_matching(condition, results):

    matched = 0
    for result in results:
        if condition(result):
            matched += 1
    return matched
```

We define the function for the combinatorial process that will be used further.

```python
from math import factorial
def combination(n,k):
  C = factorial(n) / (factorial(k) * factorial(n - k))
  return C
combination(4,2)
```

We define the function for the permutation process that will be used further.

```python
from math import factorial
def permutation(n,k):
  p = factorial(n) / (factorial(n - k))
  return p
permutation(4,2)
```

# Bayesian Theorem

**Bayes' Theorem**

For example, suppose the probability of the weather being cloudy is 40%.

Also suppose the probability of rain on a given day is 20%.

Also suppose the probability of clouds on a rainy day is 85%.

If it’s cloudy outside on a given day, what is the probability that it will rain that day?

```python
#define function for Bayes' theorem
def bayesTheorem(pA, pB, pBA):
    return pA * pBA / pB

#define probabilities
pRain = 0.2
pCloudy = 0.4
pCloudyRain = 0.85

#use function to calculate conditional probability
bayesTheorem(pRain, pCloudy, pCloudyRain)

```

```python
# Prior probability (initial belief) of a coin being fair (heads=tails=50%)
prior_fair = 0.5

# Likelihood of observing heads given a fair coin (remains 0.5)
likelihood_fair_heads = 0.5  # Assuming observation aligns with prior for a fair coin

# Update posterior probability (revised belief) considering the observation (heads)
posterior_fair = (prior_fair * likelihood_fair_heads) / (prior_fair * likelihood_fair_heads + (1-prior_fair) * (1-likelihood_fair_heads))

print("Posterior Probability of Fair Coin (after observing heads):", posterior_fair)
```

# Empirical Probability

```python
# Simulate flipping a coin 100 times
import random

flips = 100
heads = 0
for _ in range(flips):
  if random.random() < 0.5:  # Assuming a fair coin (50% chance of heads)
    heads += 1

# Calculate empirical probability of heads
empirical_prob_heads = heads / flips
print("Empirical Probability of Heads:", empirical_prob_heads)
```

# Transition Probability

```python
# Define transition probabilities for weather states (Sunny, Rainy, Cloudy)
transition_matrix = {
  "Sunny": {"Rainy": 0.3, "Cloudy": 0.4, "Sunny": 0.3},
  "Rainy": {"Sunny": 0.2, "Cloudy": 0.6, "Rainy": 0.2},
  "Cloudy": {"Sunny": 0.5, "Rainy": 0.3, "Cloudy": 0.2},
}

# Starting state (example)
current_state = "Sunny"

# Simulate a transition (choose a random next state based on probabilities)
import random
next_state = random.choices(list(transition_matrix[current_state].keys()), weights=list(transition_matrix[current_state].values()))[0]

print("Current State:", current_state)
print("Transitioned to:", next_state)
```

```python
def two_sum(nums, target):
    values = {}
    for i,num in enumerate(nums):
        if target - num in values:
            return i, values[target-num]
        values[num] = i
    return -1

print(two_sum([2,7,11,15], 9))
```

```python
def majority_element(nums):
    """
    Args:
     nums(list_int32)
    Returns:
     int32
    """
    counts = {}
    target = int(len(nums)/2)
    for num in nums:
        if num in counts:
            counts[num]+= 1
        else:
            counts[num] =1
        if counts[num] >= target:
            return num
    return 0

print(majority_element([3, 3, 3, 2, 2, 2, 3]))
```

```python
def get_intersection(numbers1, numbers2):
    """
    Args:
     numbers1(list_int32)
     numbers2(list_int32)
    Returns:
     list_int32
    """
    # Write your code here.
    sorted_intersection = []
    included_set = set()
    pointer1, pointer2 = 0,0
    while pointer1 < len(numbers1) and pointer2 < len(numbers2):
        if numbers1[pointer1] == numbers2[pointer2]:
            if numbers1[pointer1] not in included_set:
                included_set.add(numbers1[pointer1])
                sorted_intersection.append(numbers1[pointer1])
            pointer1 += 1
            pointer2 += 1
        elif numbers1[pointer1] < numbers2[pointer2]:
            pointer1 += 1
        else:
            pointer2 += 1
    
    if len(sorted_intersection) == 0:
        return [-1]
    
    return sorted_intersection

print(get_intersection([1, 2, 2, 6, 7], [2, 2, 7, 7]))
```

```python
def single_number(arr):
    """
    Args:
     arr(list_int32)
    Returns:
     int32
    """
    # Write your code here.
    visited = set()
    for num in arr:
        if num in visited:
            visited.remove(num)
        else:
            visited.add(num)

    return visited.pop()
```

```python
def generate_sorted_array_of_squares(numbers):
    """
    Args:
     numbers(list_int32)
    Returns:
     list_int32
    """
    # Write your code here.
    return list(map(lambda x: x*x, numbers))

print(generate_sorted_array_of_squares([1,2,3,4]))
```

```python
def find_maximum_sum_subarray(numbers):
    """
    Args:
     numbers(list_int32)
    Returns:
     int32
    """
    # Write your code here.
    cur_sum,max_sum = numbers[0], numbers[0]
    current_end = 1
    while(current_end <len(numbers)):
        cur_sum += numbers[current_end]
        max_sum = max(cur_sum, max_sum)
        if cur_sum <=0:
            cur_sum = 0
        current_end += 1
    return max_sum

print(find_maximum_sum_subarray([2, -6, 3, 4, -5]))
```

```python
def find_winner(votes):
    """
    Args:
     votes(list_str)
    Returns:
     str
    """
    # Write your code here.

    voter_count = {}
    max_vote_count = 0
    max_voter = None
    for vote in votes:
        if vote in voter_count:
            voter_count[vote] += 1
        else:
            voter_count[vote] = 1
        max_vote_count = max(voter_count[vote], max_vote_count)
        if voter_count[vote] == max_vote_count:
            if max_voter is None or vote < max_voter:
                max_voter = vote
            
    return max_voter

print(find_winner(["sam", "john", "sam", "john"]))
```

# Uniform Distribution
> **Scenario**: Modeling the distribution of measurement times throughout a 24-hour period.

> **Description**: Measurement times are assumed to be uniformly distributed between midnight (0 hours) and the next midnight (24 hours).

> **Data Generation**: uniform.rvs(low_uniform, high_uniform - low_uniform, size=1000) generates 1000 samples of measurement times.

> **Visualization**: Histogram of measurement times to show the uniform distribution.

```python

# Uniform Distribution
# Scenario: Measurement times throughout a 24-hour period
low_uniform = 0  # 0 hours (midnight)
high_uniform = 24  # 24 hours (next midnight)
measurement_times = uniform.rvs(low_uniform, high_uniform - low_uniform, size=1000)

# Plot Uniform distribution
plt.hist(measurement_times, bins=24, density=True, alpha=0.6, color='b', edgecolor='black')
plt.title('Uniform Distribution of Measurement Times')
plt.xlabel('Time of Day (hours)')
plt.ylabel('Density')

print(f"\nUniform Distribution: low={low_uniform}, high={high_uniform}")
print(f"Mean of measurement times: {np.mean(measurement_times)}, Range: ({min(measurement_times)}, {max(measurement_times)})")
print('\n')
```

# Beta Distribution

> **Scenario**: Modeling the proportion of days with good air quality (AQI below a certain threshold) within a given month.

> **Description**: The proportion of days with good air quality can be modeled using a Beta distribution, which is suitable for modeling proportions.

> **Data Generation**: beta.rvs(a, b, size=1000) generates 1000 samples of the proportion of days with good air quality, where a and b are the shape parameters of the Beta distribution.

> **Visualization**: Histogram of the proportions of days with good air quality, overlaid with the probability density function (PDF) of the Beta distribution.


You can also use this [link](https://drive.google.com/file/d/1XvC_LYa6CS7TyA32LZWDNKKj7TipohSr/view?usp=sharing) to enhance your understanding of beta distributions.

```python
from scipy.stats import beta

# Beta Distribution
# Scenario: Proportion of days with good air quality (AQI below a certain threshold)
a, b = 2, 5  # Shape parameters for the Beta distribution
beta_samples = beta.rvs(a, b, size=1000)

# Plot Beta distribution
plt.hist(beta_samples, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = beta.pdf(x, a, b)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Beta Distribution of Proportion of Good Air Quality Days')
plt.xlabel('Proportion of Good Air Quality Days')
plt.ylabel('Density')
plt.show()

# Additional Analysis and Summary
print("\n")
print(f"Beta Distribution: a={a}, b={b}")
print(f"Mean of Beta samples: {np.mean(beta_samples)}, Std of Beta samples: {np.std(beta_samples)}")
print("\n")
```

# Normal Distribution

> **Scenario**: Modeling the distribution of daily PM2.5 (particulate matter) concentrations.

> **Description**: PM2.5 concentration levels typically follow a normal distribution with a given mean and standard deviation.

> **Data Generation**: norm.rvs(mean_pm25, std_pm25, size=1000) generates 1000 samples of PM2.5 concentrations.

> **Visualization**: Histogram of the PM2.5 samples overlaid with the probability density function (PDF) of the normal distribution.

You can refer to this [link](https://colab.research.google.com/drive/1XvC_LYa6CS7TyA32LZWDNKKj7TipohSr#scrollTo=DKIppbfKGpiW) for understanding normal distribution in detail.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, multivariate_normal

# Normal Distribution
# Scenario: Daily PM2.5 concentration levels
mean_pm25 = 50  # Mean PM2.5 concentration
std_pm25 = 10   # Standard deviation of PM2.5 concentration
pm25_samples = norm.rvs(mean_pm25, std_pm25, size=1000)

# Plot Normal distribution
plt.hist(pm25_samples, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_pm25, std_pm25)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution of PM2.5 Concentration')
plt.xlabel('PM2.5 Concentration (µg/m³)')
plt.ylabel('Density')

# Additional Analysis and Summary
print(f"Normal Distribution: mean={mean_pm25}, std={std_pm25}")
print(f"Mean of PM2.5 samples: {np.mean(pm25_samples)}, Std of PM2.5 samples: {np.std(pm25_samples)}")
print("\n")
```

#Multivariate Normal Distribution
> **Scenario**: Modeling the relationship between temperature, humidity, and PM2.5 concentrations.

> **Description**: Temperature, humidity, and PM2.5 levels can be modeled using a multivariate normal distribution.

> **Data Generation**: multivariate_normal.rvs(mean_multivariate, cov_multivariate, size=1000) generates 1000 samples of temperature, humidity, and PM2.5 levels.

> **Visualization**: Scatter plots to visualize the relationships between temperature and humidity, and between humidity and PM2.5 levels.

You can explore this [notebook](https://colab.research.google.com/drive/1XvC_LYa6CS7TyA32LZWDNKKj7TipohSr#scrollTo=JViUjI5mK28K&line=3&uniqifier=1) later.



```python
# Multivariate Normal Distribution
# Scenario: Temperature, Humidity, and PM2.5 concentrations
mean_multivariate = [20, 50, 50]  # Mean temperature (°C), humidity (%), PM2.5 (µg/m³)
cov_multivariate = [
    [10, 5, 2],    # Covariance matrix
    [5, 20, 5],
    [2, 5, 15]
]
multivariate_samples = multivariate_normal.rvs(mean_multivariate, cov_multivariate, size=1000)

# Plot Multivariate Normal distribution
plt.subplot(1,2,1)
plt.scatter(multivariate_samples[:, 0], multivariate_samples[:, 1], alpha=0.6, color='r')
plt.title('Multivariate Normal Distribution')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')

plt.subplot(1,2,2)
plt.scatter(multivariate_samples[:, 1], multivariate_samples[:, 2], alpha=0.6, color='purple')
plt.title('Multivariate Normal Distribution')
plt.xlabel('Humidity (%)')
plt.ylabel('PM2.5 Concentration (µg/m³)')

plt.tight_layout()
plt.show()

print("\n")
print(f"\nMultivariate Normal Distribution: mean={mean_multivariate}")
print(f"Mean of samples: {np.mean(multivariate_samples, axis=0)}")
print(f"Covariance matrix of samples:\n{np.cov(multivariate_samples, rowvar=False)}")
print("\n")
```

# t Distribution

> **Scenario**: Analyzing the differences in average PM2.5 concentrations between weekdays and weekends to account for the smaller sample size.

> **Description**: The differences in average PM2.5 concentrations between weekdays and weekends can be modeled using a t distribution, which is useful for small sample sizes.

> **Data Generation**: t.rvs(df, size=1000) generates 1000 samples of the difference in average PM2.5 concentrations, where df is the degrees of freedom.

> **Visualization**: Histogram of the differences in average PM2.5 concentrations, overlaid with the probability density function (PDF) of the t distribution.

```python
from scipy.stats import t

# t Distribution
# Scenario: Differences in average PM2.5 concentrations between weekdays and weekends
df = 10  # Degrees of freedom for the t distribution
t_samples = t.rvs(df, size=1000)

# Plot t distribution
plt.hist(t_samples, bins=30, density=True, alpha=0.6, color='r', edgecolor='black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = t.pdf(x, df)
plt.plot(x, p, 'k', linewidth=2)
plt.title('t Distribution of Differences in Average PM2.5 Concentrations')
plt.xlabel('Difference in Average PM2.5 Concentrations')
plt.ylabel('Density')
plt.show()

# Additional Analysis and Summary
print("\n")
print(f"t Distribution: df={df}")
print(f"Mean of t samples: {np.mean(t_samples)}, Std of t samples: {np.std(t_samples)}")
print("\n")
```

### 1. Call Center Analysis
<image src="https://img.freepik.com/free-vector/flat-customer-support-illustration_23-2148899114.jpg?size=338&ext=jpg&ga=GA1.1.2113030492.1720051200&semt=ais_user" />

You are analyzing the number of customer service calls received at a call center, which follows a Poisson distribution with an average of 5 calls per hour. Calculate the probability of receiving exactly 3 calls in the next hour.

**Instructions:**
Write Python code to compute the probability of a specific number of events (calls) in a Poisson distribution. Plot the probability mass function (PMF) and cumulative distribution function (CDF) for visualization.

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt

# Parameters
lambda_ = 5  # Average number of calls per hour
k = 3  # Number of calls we are interested in

# Probability of exactly 3 calls in the next hour
prob_3_calls = poisson.pmf(k, lambda_)
print(prob_3_calls)

# Plotting PMF
x = np.arange(0, 15)
pmf = poisson.pmf(x, lambda_)
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, align='center', alpha=0.6)
plt.title('Probability Mass Function (PMF) of Customer Service Calls')
plt.xlabel('Number of Calls')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(True)
plt.show()

# Plotting CDF
cdf = poisson.cdf(x, lambda_)
plt.figure(figsize=(10, 6))
plt.step(x, cdf, where='mid', label='CDF')
plt.title('Cumulative Distribution Function (CDF) of Customer Service Calls')
plt.xlabel('Number of Calls')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid(True)
plt.show()

# Print results
print(f"Probability of exactly 3 calls in the next hour: {prob_3_calls:.4f}")
```

### 2. Conversion rate of visitors on a website
<image src="https://img.freepik.com/free-vector/business-intelligence-experts-transform-data-into-useful-information-business-intelligence-business-analysis-it-management-tools-concept_335657-1889.jpg?size=626&ext=jpg&ga=GA1.1.2082370165.1716508800&semt=ais_user" width=500/>

You are analyzing the conversion rate of visitors on a website. The conversion rate is 10%. Calculate the probability that exactly 2 out of 10 randomly selected visitors make a purchase.

**Instructions:**
Write Python code to compute the probability of a specific number of successes (visitors making a purchase) in a Bernoulli distribution. Plot the probability mass function (PMF) and cumulative distribution function (CDF) for visualization.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

# Parameters
n = 10  # Number of trials (visitors)
p = 0.1  # Probability of success (visitor making a purchase)
k = 2  # Number of successes we are interested in

# Probability of exactly 2 visitors making a purchase
prob_2_purchases = binom.pmf(k, n, p)

# Plotting PMF
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, align='center', alpha=0.6)
plt.title('Probability Mass Function (PMF) of Website Visitor Purchases')
plt.xlabel('Number of Visitors Making a Purchase')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(True)
plt.show()

# Plotting CDF
cdf = binom.cdf(x, n, p)
plt.figure(figsize=(10, 6))
plt.step(x, cdf, where='mid', label='CDF')
plt.title('Cumulative Distribution Function (CDF) of Website Visitor Purchases')
plt.xlabel('Number of Visitors Making a Purchase')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid(True)
plt.show()

# Print results
print(f"Probability of exactly 2 visitors making a purchase: {prob_2_purchases:.4f}")
```

### 3. Pass rate of students in a certification exam

<img src="https://img.freepik.com/free-vector/college-entrance-exam-concept-illustration_114360-10202.jpg?size=338&ext=jpg&ga=GA1.1.2113030492.1720051200&semt=ais_user" />

You are analyzing the pass rate of students in a certification exam. Historically, 70% of students pass the exam. Calculate the probability that at least 5 out of 8 randomly selected students pass the exam.

**Instructions:**
Write Python code to compute the probability of at least a specific number of successes (students passing the exam) in a binomial distribution. Plot the probability mass function (PMF) and cumulative distribution function (CDF) for visualization.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

# Parameters
n = 8  # Number of trials (students)
p = 0.7  # Probability of success (student passing the exam)
k = 5  # Minimum number of successes we are interested in

# Probability of at least 5 students passing the exam
# this is where we make use CDF - for atleast use cases. 
prob_at_least_5_pass = 1 - binom.cdf(k-1, n, p)

# Plotting PMF
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, align='center', alpha=0.6)
plt.title('Probability Mass Function (PMF) of Exam Passes')
plt.xlabel('Number of Students Passing the Exam')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(True)
plt.show()

# Plotting CDF
cdf = binom.cdf(x, n, p)
plt.figure(figsize=(10, 6))
plt.step(x, cdf, where='mid', label='CDF')
plt.title('Cumulative Distribution Function (CDF) of Exam Passes')
plt.xlabel('Number of Students Passing the Exam')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid(True)
plt.show()

# Print results
print(f"Probability of at least 5 students passing the exam: {prob_at_least_5_pass:.4f}")
```
